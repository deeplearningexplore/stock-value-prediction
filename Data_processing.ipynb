```python
csv_files = Path(path).glob('*.csv')
l = []
gr = pd.DataFrame()

for i in csv_files:
  df = pd.read_csv(i)
  # df2 = df.loc[(df["Date"] >= '2017-01-01')]
  df['Date'] = pd.to_datetime(df['Date'])
  df2 = df[(df['Date'] > "2017-01-01")]
  df2 = df2[["Date","Close"]]
  name = os.path.splitext(i)[0]
  df2.rename(columns= {"Close":name[53:]}, inplace= True)
  df2.set_index('Date', inplace = True)
  l.append(df2)
l
  
```

```python
gr = pd.concat(l,join = 'outer', ignore_index = False, axis =1)
```
