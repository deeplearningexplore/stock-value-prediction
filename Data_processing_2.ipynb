```python
df = pd.read_csv("/content/gdrive/MyDrive/Colab Notebooks/weather.csv")
```

```python
df.head()
```

```python
df2 = df[["name","datetime","temp","humidity"]]
```

```python
groups = [y for x, y in df2.groupby('name')]
```

```python
df_cal = groups[0]
df_cal= df_cal.drop("name",axis=1)
df_cal = df_cal.rename(columns={"temp":"cal_temp", "humidity":"cal_hum","datetime":"Date"})
df_cal.set_index('Date', inplace = True)
# df_cal.set_index("datetime",inplace = True)

df_flo = groups[1]
df_flo = df_flo.drop("name",axis=1)
df_flo = df_flo.rename(columns={"temp":"flo_temp", "humidity":"flo_hum","datetime":"Date"})
df_flo.set_index('Date', inplace = True)

df_may = groups[2]
df_may = df_may.drop("name",axis=1)
df_may = df_may.rename(columns={"temp":"may_temp", "humidity":"may_hum","datetime":"Date"})
df_may.set_index('Date', inplace = True)

df_new = groups[3]
df_new = df_new.drop("name",axis=1)
df_new = df_new.rename(columns={"temp":"new_temp", "humidity":"new_hum","datetime":"Date"})
df_new.set_index('Date', inplace = True)

df_tex = groups[4]
df_tex = df_tex.drop("name",axis=1)
df_tex = df_tex.rename(columns={"temp":"tex_temp", "humidity":"tex_hum","datetime":"Date"})
df_tex.set_index('Date', inplace = True)
```

```python

df_cal.head()
```

```python
df_weather = pd.concat([df_cal,df_may,df_new,df_flo,df_tex],join = 'outer', ignore_index = False, axis =1)
```
